```python
def top10_channels(i,count,c):
    plt.subplot(2,2,i)
    plt.title("Top 10 Channels with most number of "+count)
    sns.barplot(x=channel.sort_values(by=count, ascending=False).head(10)[count],
                y=channel.sort_values(by=count, ascending=False).head(10)['channelTitle'],
                color=c, saturation=1)   

plt.figure(figsize=(23,11))
top10_channels(1,'view_count','r')
top10_channels(2,'comment_count','b')
top10_channels(3,'likes','g')
top10_channels(4,'dislikes','y')
```

```python
category = df[['Category','view_count','comment_count','likes','dislikes']].groupby(by='Category').sum().reset_index()
category.head()
```

```python
def top10_categories(i,count,c):  
    plt.subplot(2,2,i)
    plt.title("Top 10 categories with most number of "+count)
    sns.barplot(x=category.sort_values(by=count, ascending=False).head(10)[count],
                y=category.sort_values(by=count, ascending=False).head(10)['Category'],
                color=c, saturation=1);
    
plt.figure(figsize=(23,11))
top10_categories(1,'view_count','r')
top10_categories(2,'comment_count','b')
top10_categories(3,'likes','g')
top10_categories(4,'dislikes','y')
```

**Usually, In most cases, viewers just watch the video and turn to other video without Commenting, Liking or Disliking.But if they responded to a video then it signifies something**

**Let's create a like%, dislike% and comment% columns which signifies the public response to a
video:**

```python
category['%comment'] = round((category['comment_count']/category['view_count'])*100, 3)
category['%likes'] = round((category['likes']/category['view_count'])*100, 3)
category['%dislikes'] = round((category['dislikes']/category['view_count'])*100, 3)

category
```

```python
category.plot(x='Category', y=['%comment','%likes','%dislikes'], kind='bar', figsize=(12,4))
plt.show()
```

### Categories with most %like, %dislike and %comment:

```python
def top_response(col,c):
    plt.figure(figsize=(12,4))
    plt.xticks(rotation=90)
    plt.title("Category with more "+col)
    sns.barplot(x=category.sort_values(by=col, ascending=False)['Category'],
                y=category.sort_values(by=col, ascending=False)[col],
                color=c, saturation=1);
    
for i,j in zip(['%likes','%dislikes','%comment'], ['r','b','g']):
    top_response(i,j)
```

## No of days took for a video to get into trending

```python
# data of a videos before entering into trending:

df1 = data.sort_values(by='view_count', ascending=True)
df1 = df1.drop_duplicates(subset='video_id', keep='first')
df1.head(2)
```

```python
df1.shape
```

```python
release = list(df1['publishedAt'])
start = list(df1['trending_date'])

# Subtracting 'publishedAt' colum from 'trending_date' column.
# This gives us the no.of.days taken for a video to go trending.
days_taken = [ start[i]-release[i] for i in range(len(start)) ]
df1['days_taken']=days_taken

df1.head(2)
```

```python
# Taking only days and ignoring time:

days_taken = []

for i in df1['days_taken']:
    a=str(i).split(' ')[0]
    days_taken.append(a)
```

```python
# '-1' is replaced with '0' as the time taken is less than a day:

days = []

for i in days_taken:
    if i=='-1':
        days.append(0)
    else:
        days.append(int(i))
```

```python
df1['days_taken'] = days
df1.head(2)
```

### No.of.days taken Vs No.of.videos

```python
count_of_days = pd.DataFrame(df1['days_taken'].value_counts()).reset_index().sort_values(by='days_taken')
count_of_days.head(10)
```

```python
plt.figure(figsize=(12,3))
plt.title("No.of.days taken Vs No.of.videos")

sns.barplot(x=count_of_days['days_taken'],
            y=count_of_days['count'],
            color='purple', saturation=1);
```

* **Nearly 40,000 videos of 46,294 videos comes to trending in lessthan a day**
* **So, we can say that average time taken for all categories and channels to get into trending is lessthan a day**
